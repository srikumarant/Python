```python

import numpy as np
from numpy.random import randn
N = 1000
counter = 0
for x in randn(N):    
    if(x >= -1 and x <= 1):
        counter = counter + 1
        print("randn of", N, "is:", x, "which is between -1 and 1. counter is:", counter)
        
print("--------------------------------------------------------------------------------")
print("Counter is:", counter)
print("N is:", N)
print("Result is (counter/N):", counter / N) 

```
